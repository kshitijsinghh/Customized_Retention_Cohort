```python
import psycopg2
import numpy as np
import pandas as pd
```

```python
ll = np.arange(32,1,-1)
```

```python
def teacher_4week_activity(start, stop, step):
    conn = psycopg2.connect(dbname = '*****', user = '*****', host = '*****', password = '*****')
    week_list = np.arange(start, stop, step)
    overall_list = []
    for num in week_list:
      if(num<4):
        break
      i=1
      curr_list = list()
      sql1 = '''SELECT   count(u.id) FROM "User" u WHERE date(u."createdAt")>=CURRENT_DATE - INTERVAL '{0}' AND date(u."createdAt") < CURRENT_DATE - INTERVAL '{1}' and u."type" = '{2}' '''.format(str(num)+"week",str(num-1)+"week", "Teacher")
      cur1 = conn.cursor()
      cur1.execute(sql1)
      result1 = cur1.fetchall()[0][0]
      curr_list.append(result1)
      # print(result1)
      up = num
      down = num -2
      while i<5:
        
        sql2 = '''SELECT count(DISTINCT(c."authorId")) FROM "Contest" c INNER JOIN "User" u on c."authorId"= u.id WHERE date(u."createdAt")>=CURRENT_DATE - INTERVAL '{0}' AND u."createdAt" < CURRENT_DATE - INTERVAL '{1}' and u."type" = '{2}' AND date(c."createdAt")>=CURRENT_DATE - INTERVAL '{3}' AND date(c."createdAt") < CURRENT_DATE - INTERVAL '{4}' '''.format(str(num)+"week",str(num-1)+"week", "Teacher",str(up)+"week",str(down)+"week")
        cur2 = conn.cursor()
        cur2.execute(sql2)
        
        result2 = cur2.fetchall()[0][0]
        # print(result2)
        curr_list.append(round((result2/result1)*100, 2))
        i= i+1
        up= up-2
        down = down-2
        
      overall_list.append(curr_list)
    conn.close()
    df = pd.DataFrame(overall_list)
    colm= ['Number Of Teachers Acquired', '%age Teachers Active in 0-2 Week','%age Teachers Active in 2-4 Week','%age Teachers Active in 4-6 Week','%age Teachers Active in 6-8 Week']
    df.columns = colm 
    df['Week Number'] = np.arange(32,3,-1)
    df.to_csv('Teachers_Activity.csv', index=False)
    return df
    





        
```

```python
teacher_4week_activity(32, 0, -1)
```

```python
def Student_4week_activity(start, stop, step): 
    conn = psycopg2.connect(dbname = '*****', user = '*****', host = '*****', password = '*****')
    week_list = np.arange(start, stop, step)
    overall_list = []
    for num in week_list:
      if(num<4):
        break
      i=1
      curr_list = list()
      sql1 = '''SELECT   count(u.id) FROM "User" u WHERE date(u."createdAt")>=CURRENT_DATE - INTERVAL '{0}' AND date(u."createdAt") < CURRENT_DATE - INTERVAL '{1}' and u."type" = '{2}' '''.format(str(num)+"week",str(num-1)+"week", "Student")
      cur1 = conn.cursor()
      cur1.execute(sql1)
      result1 = cur1.fetchall()[0][0]
      curr_list.append(result1)
      # print(result1)
      up = num
      down = num -2
      while i<5:
        
        sql2 = '''SELECT count(DISTINCT(cp."studentId")) FROM "ContestParticipants" cp INNER JOIN "User" u on cp."studentId"= u.id WHERE date(u."createdAt")>=CURRENT_DATE - INTERVAL '{0}' AND u."createdAt" < CURRENT_DATE - INTERVAL '{1}' and u."type" = '{2}' AND date(cp."date")>=CURRENT_DATE - INTERVAL '{3}' AND date(cp."date") < CURRENT_DATE - INTERVAL '{4}'  '''.format(str(num)+"week",str(num-1)+"week", "Student",str(up)+"week",str(down)+"week")
        cur2 = conn.cursor()
        cur2.execute(sql2)
        
        result2 = cur2.fetchall()[0][0]
        # print(result2)
        curr_list.append(round((result2/result1)*100, 2))
        i= i+1
        up= up-2
        down = down-2
        
      overall_list.append(curr_list)
    conn.close()
    df = pd.DataFrame(overall_list)
    colm= ['Number Of Students Acquired', '%age Students Active in 0-2 Week','%age Students Active in 2-4 Week','%age Students Active in 4-6 Week','%age Students Active in 6-8 Week']
    df.columns = colm 
    df['Week Number'] = np.arange(32,3,-1)
    df.to_csv('Student_Activity.csv', index=False)
    return df
    
```
